# Python Basics: NaNs and Statistics #

```python
# Import Libraries
import pandas as pd
```

```python
# Import files

# Define a path
path = r'..\additional_data\stocks_returns.csv'

# Read the CSV
dataframe = pd.read_csv(path)
dataframe.set_index('Date', inplace=True)
dataframe.index = pd.to_datetime(dataframe.index)

# Show it
dataframe
```

### Missing Values ###

```python
# Detect NaNs
dataframe.isna().sum().sort_values(ascending=False)
```

```python
# Drop Missing Values
dataframe.dropna() # This is an empty dataframe. Why?
```

```python
# Drop Rows with Any
dataframe.dropna(axis=0, how="any")
```

```python
# Drop Columns with Any
dataframe.dropna(axis=1, how="any")
```

```python
# Drop Rows if All
dataframe.dropna(axis=0, how="all")
```

```python
# Drop Columns if All
dataframe.dropna(axis=1, how="all")
```

```python
# Drop with a Threshold
dataframe.dropna(axis=0, thresh=500)  # keep rows with at least 2 non-NaN
```

```python
# Using Subset
drop_subset = dataframe.dropna(subset=["ABNB", "UBER"])  # drop rows where A or B is NaN

drop_subset
```

```python
# Fill Missing Values
ffill_df = dataframe.ffill()
ffill_df
```

```python
# Fill Missing Values
bfill_df = dataframe.bfill()
bfill_df
```

```python
# Fill with Any Value
zfill_df = dataframe.fillna(0)
zfill_df
```

```python
# Fill with mean
fill_mean = dataframe.fillna(dataframe.mean(numeric_only=True))
fill_mean
```

### Statistics ###

```python
# Choose a stock
stock = dataframe['AAPL']
```

```python
# -------------------------------
# 1. CENTRAL TENDENCY MEASURES
# -------------------------------
print("\n--- Central Tendency ---")
print("Mean Returns:", stock.mean())
print("Median Returns:", stock.median())
print("Mode Returns:", stock.mode()[0])  # mode returns a series
```

```python
# -------------------------------
# 2. DISPERSION MEASURES
# -------------------------------
print("\n--- Dispersion ---")
print("Returns Standard Deviation:", stock.std())
print("Returns Variance:", stock.var())
print("Returns Range:", stock.max() - stock.min())
```

```python
# -------------------------------
# 3. CORRELATION
# -------------------------------
print("\n--- Correlation ---")
correlation_matrix = dataframe[['AAPL', 'META', 'NVDA', 'TSLA']].corr()
print(correlation_matrix)
print("Correlation AAPL vs NVDA:", dataframe["AAPL"].corr(dataframe["NVDA"]))
```

```python
# -------------------------------
# 4. DESCRIPTIVE SUMMARY
# -------------------------------
print("\n--- Descriptive Summary ---")
print(dataframe[['AAPL', 'META', 'NVDA', 'TSLA']].describe())
```

```python
# -------------------------------
# 5. APPLY BASIC STATISTICS USING APPLY
# -------------------------------
# Create a custom function
def z_score(x):
    return (x - x.mean()) / x.std()

df_z = dataframe[['AAPL', 'META', 'NVDA', 'TSLA']].apply(z_score)
print("\nZ-scores for each column:")

df_z.dropna()
```
