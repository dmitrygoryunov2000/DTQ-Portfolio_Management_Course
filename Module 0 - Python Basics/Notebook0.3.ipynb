# Python Basics: Loops, Functions and More #

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##############################################################
# 1. WHILE LOOPS
##############################################################
# A while loop runs until a certain condition is no longer True.

x = 0
while x < 20:
    x = x + 2
    print(x)
```

```python
##############################################################
# 2. CONDITIONALS INSIDE LOOPS
##############################################################
# We can use if / elif / else to check conditions inside a loop.

x = 0
while x < 20:
    if (x % 2) == 0:
        print("Number " + str(x) + " is multiple of 2")
    elif (x % 3) == 0:
        print("Number " + str(x) + " is multiple of 3")
    else:
        print("Not multiple of 2 or 3")
    x += 1
```

```python
##############################################################
# 3. LOGICAL OPERATORS (and, or, not)
##############################################################
# Combine conditions using and, or, not.

x = 0
while x < 20:
    if (x % 2 == 0) and (x % 3 == 0):
        print("Number " + str(x) + " is multiple of 2 AND 3")
    elif (x % 2 == 0) or (x % 3 == 0):
        print("Number " + str(x) + " is multiple of 2 OR 3")
    else:
        print("Not divisible by 2 or 3")
    x += 1
```

```python
##############################################################
# 4. LOOP CONTROLS: pass, continue, break
##############################################################
# - pass: does nothing (placeholder)
# - continue: skips current iteration
# - break: exits the loop completely

x = 0
while x < 10:
    x += 1
    if x == 3:
        pass  # nothing happens
    elif x == 5:
        continue  # skips printing 5
    elif x == 8:
        break  # loop stops when x == 8
    print("x =", x)
```

```python
##############################################################
# 5. ERROR HANDLING (try, except, finally)
##############################################################
# Handle errors gracefully.

try:
    num = int("hello")  # will raise ValueError
except ValueError as e:
    print("An error occurred:", e)
finally:
    print("This block always executes.")
```

```python
##############################################################
# 6. FOR LOOPS
##############################################################
# For loops are used to iterate over sequences (lists, strings, ranges, etc.)

# (a) Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print("Fruit:", fruit)
```

```python
# (b) Iterating over a string
for char in "Python":
    print("Letter:", char)
```

```python
# (c) Using range() to generate numbers
for i in range(5):  # 0 to 4
    print("Number:", i)
```

```python
# (d) Iterating with indexes using enumerate()
for index, fruit in enumerate(fruits):
    print("Index:", index, "Fruit:", fruit)
```

```python
# (e) Iterating over dictionary items
student = {"name": "Edgar", "age": 25, "major": "Economics"}
for key, value in student.items():
    print(key, ":", value)
```

```python
# (f) Nested loops
for i in range(1, 4):
    for j in range(1, 3):
        print(f"i={i}, j={j}")
```

```python
##############################################################
# 7. FUNCTIONS
##############################################################

# (a) Basic functions
def add_numbers(a, b):
    """Return the sum of two numbers."""
    return a + b

print("Sum:", add_numbers(2, 3))
```

```python
# (b) Function with default parameter
def greet(name="Anonymous"):
    """Print a greeting with a default name."""
    print("Hello,", name)

greet()
greet("Edgar")
```

```python
# (c) Function using a DataFrame
def column_mean(df, col):
    """Return the mean of a given column in a DataFrame."""
    return df[col].mean()

df = pd.DataFrame({"values": [10, 20, 30, None, 40]})
print("Mean of 'values':", column_mean(df, "values"))
```

```python
##############################################################
# 8. MINI UTILITIES
##############################################################

# (a) String cleaning
def pretty(text):
    """Remove leading and trailing spaces and convert to lowercase."""
    return text.strip().lower()

print(pretty("   HeLLo   "))
```

```python
# (a) String cleaning
def beauty(text):
    """Capitalize the first letter of each word."""
    return text.title()

print(beauty("python is great"))
```

```python
# (b) Word splitting
def split_words(text):
    """Split text into words."""
    return text.split()

print(split_words("Python is fun"))
```

```python
##############################################################
# 9. COLLATZ CONJECTURE
##############################################################

def collatz_conjecture(integer): 
    """Generate the Collatz series starting from an integer."""
    series = [integer]

    while integer != 1:
        if integer % 2 == 0:
            integer = integer // 2  
        else:  # odd number
            integer = 3 * integer + 1
        series.append(integer) 
            
    return series

# Example run
n = 19
sequence = collatz_conjecture(n)
print("Collatz sequence for", n, ":", sequence)
```

```python
##############################################################
# 10. VISUALIZATION OF THE SERIES
##############################################################

plt.figure(figsize=(8, 5))
plt.plot(range(len(sequence)), sequence, marker="o", linestyle="-", color="b")

plt.title(f"Collatz Conjecture Sequence starting at {n}")
plt.xlabel("Step")
plt.ylabel("Value")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
